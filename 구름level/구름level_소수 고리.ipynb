```python
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
from itertools import permutations

user_input = int(input())

def IsPrime(a, b):
	x = a + b
	for i in range(2, x):
		if x%i == 0:
			return False
	return True
		
for i in permutations(range(2, user_input+1), user_input-1):
	tmp = [1]+list(i)
	flag = True
	for j in range(len(tmp)):
		if not IsPrime(tmp[j], tmp[j-1]):
			flag = False
			break
			
	if flag:
		result = ''
		for k in tmp:
			result += f'{k} '
		print(result[:-1])
```
