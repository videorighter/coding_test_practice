```python
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean

from itertools import permutations
user_input = input()
print ("Hello Goorm! Your input is " + user_input)
```

```python
def IsPrime(num):
    for i in range(2, num, 1):
        if num % i == 0:
            return False
        else:
            return True
```

```python
arr = range(2, int(user_input)+1)
candidates = permutations(arr, len(arr))
for i in candidates:
    tmp = set()
    tmp.add(1+i[0])
    tmp.add(i[-1]+1)
    for idx in range(1, len(i)):
        tmp.add(i[idx]+i[idx-1])
        flag = True
    for num in tmp:
        if not IsPrime(num):
            flag = False
            break
    if flag:
        print("1 " + " ".join(map(str, i)))
```
